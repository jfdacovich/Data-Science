````python
import random

value = random.randint(1,6)
count = 0
guess = 0

while guess != value:
    count += 1 #incremento
    guess = input("Adivina un numero entre 1 y 6: ")
    if guess.isnumeric ():
        guess = int(guess)
        if guess > 6:
            print("por favor inserta un numero menor o igual a 6")
        if guess < value:
            print("Tu numero es muy bajo")
        if guess > value:
            print("Tu numero es muy alto")
        if guess == value:
            break
    elif guess.isalpha():
        print("Por favor introduce un numero")
            
if guess == value:
    count = str(count)
    print ("Genial has adivinado el numero en " + count + " intentos" )
    
````

````python
# Ejemplo datos

import random

names = [] # variable global

for name in range(5,1000,10):
    names.append(name) #añadimos cada valor a la variable "name"
print (names)
````

````python
for name in range(100):
    names.append(name)
    print (names)
````

````python
for name in range (100):
    names.append(name)
print(random.choices(names,k=5))
````

````python
#utilizas seed para limitar y/o acotar probabilidades dentro de un algoritmo
random.seed(666)
names = []
for name in range (100):
    names.append(name)
print(random.choices(names,k=5))
````
