# Piedra, papel, tijeras

Vamos a jugar contra nuestro ordenador al famoso juego. 

[Wikipedia Piedra, Papel o Tijeras](https://es.wikipedia.org/wiki/Piedra,_papel_o_tijera)

Se aconseja el uso de funciones

## Objetivos
1. Uso de bucle
2. Captura de datos por consola
3. Uso if-elif-else
4. Uso de try-except
5. Definición de funciones. Programación modular. 
6. Operadores lógicos. 
7. Imprimir por consola
8. Importación de módulos externos

```python
# Importaremos la función choice del módulo random
import random
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

# Asignamos a una lista las 3 posibles opciones: 'piedra', 'papel' o 'tijeras'. 
opciones = ["piedra", "papel", "tijeras"]

# Asignamos una variable al número de partidas máxima: 1, 3, 5, etc...
partidas = int(input("Introduce la cantidad de partidas que deseas jugar: "))

# Asignamos una variable al número de partidas que debe ganar un jugador para ganar. 
partidas_para_ganar = 0

if partidas == 3:
    partidas_para_ganar = 2
elif partidas ==5:
    partidas_para_ganar = 3
else:
    partidas_para_ganar = 1
    
# Las siguientes variables acumularan las partidas ganadas de cada participante
pc_win = 0
jugador_win = 0
empate = 0

# Definimos una función que devuelva aleatoriamente una de las 3 opciones.
# Esto corresponderá a la jugada de la máquina. Totalmente aleatoria.
def cpu_random():
    pc_random = random.choice(opciones) 
    return (pc_random) 
 
# Definimos una función que pregunte tu elección: 'piedra', 'papel' o 'tijeras'
def opciones_a_escoger():
    opcion_a_escoger = input ("Escoge una de las siguientes opciones: piedra, papel o tijeras! ")
    while opcion_a_escoger not in opciones:
        opcion_a_escoger = input ("Recuerda escoger piedra, papel , o tijeras ")
    return opcion_a_escoger

# Definimos una función que resuelva un combate. 
# Devuelve 0 si hay empate, 1 si gana la máquina, 2 si gana el jugador humano 
def resultado(opcion_a_escoger, pc_random):
    global pc_win
    global jugador_win
    global empate
    if opcion_a_escoger == "piedra" and pc_random == "tijeras":
        jugador_win +=1
        return (2)
    elif opcion_a_escoger == "tijeras" and pc_random == "papel":
        jugador_win +=1
        return (2)
    elif opcion_a_escoger == "papel" and pc_random == "piedra":
        jugador_win +=1
        return (2)
    elif opcion_a_escoger == "piedra" and pc_random == "papel":
        pc_win +=1
        return (1)
    elif opcion_a_escoger == "tijeras" and pc_random == "piedra":
        pc_win +=1
        return (1)
    elif opcion_a_escoger == "papel" and pc_random == "tijeras":
        pc_win +=1
        return (1)
    else:
        empate +=1
        return(0)
    
# Definimos una función que muestre la elección de cada jugador y el estado de la partida
# Esta función se utilizara cada vez que se actualicen los puntos acumulados
def estatus ():
    print(f"El jugador ha elegido {opcion_a_escoger}")
    print(f"La maquina ha elegido {pc_random}")
    print("----------------------------------------")
    print(f"El jugador lleva {jugador_win} partidas ganadas")
    print(f"La maquina lleva {pc_win} partidas ganadas")
    print("----------------------------------------")
 

# Creamos un bucle while que itere mientras ningún jugador alcance el mínimo de victorias necesarias para ganar. 
# Dentro del bucle se resuelve la jugada de la máquina y pregunta la del jugador. 
# Se comparan y se actualizan los valores acumulados de las variables ganadas por cada participante

while jugador_win < partidas_para_ganar and pc_win < partidas_para_ganar and empate + jugador_win + pc_win < partidas:
    pc_random = cpu_random()
    opcion_a_escoger = opciones_a_escoger()
    resultado(opcion_a_escoger, pc_random)
    estatus()
    
# Se anuncia por consola el ganador del juego en función de quién tiene más victorias 
# aculumadas
    if jugador_win > pc_win and jugador_win > empate:
        print("El jugador tiene mas victorias acumuladas")
    elif pc_win > jugador_win and pc_win > empate:
        print("Estas siendo vencido por una maquina")
        print("La maquina acumula mas victorias")
    elif empate> jugador_win and empate > pc_win:
        print("tenemos un empate de momento")
```
